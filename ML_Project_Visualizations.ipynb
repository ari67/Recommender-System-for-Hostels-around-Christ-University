## OVERALL RATING OF THE HOSTELS AND CURFEW TIMING

```python
# Create a contingency table
ct = pd.crosstab(data['Curfew (By what time you have to reach hostel)'], data['On a scale of 1 to 5, select the overall rating'])

# Plot heatmap of the contingency table
sns.heatmap(ct, annot=True, cmap='YlGnBu')

# Set plot title and axis labels
plt.title("Overall Rating vs Curfew")
plt.xlabel("Overall Rating")
plt.ylabel("Curfew")

# Show the plot
plt.show()
```

```python
# Create a cross tabulation of the two categorical variables
cross_tab = pd.crosstab(data["On a scale of 1 to 5, select the overall rating"], data["Curfew (By what time you have to reach hostel)"])

# Plot the stacked bar chart
cross_tab.plot(kind="bar", stacked=True, figsize=(8,6))

# Set plot title and axis labels
plt.title("Overall Rating vs Curfew")
plt.xlabel("Overall Rating")
plt.ylabel("Number of Hostels")

# Show the plot
plt.show()
```

## CHI-SQUARE TEST

```python

from scipy.stats import chi2_contingency

# Create a contingency table of the two categorical variables
contingency_table = pd.crosstab(data['On a scale of 1 to 5, select the overall rating'],
                                data['Curfew (By what time you have to reach hostel)'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results of the chi-square test
print("Chi-square statistic:", chi2)
print("p-value:", p)
```

### Here p-value is 0.3351 which is greater than 0.05 therefore we fail to reject the null hypothesis, ie there is no significant relationship between the two categorical variables overall ratings and curfew timing

## OVERALL RATING VS DISTANCE FROM COLLEGE

```python
# Create a cross-tabulation of the two variables
ct = pd.crosstab(data["On a scale of 1 to 5, select the overall rating"],
                 data["Distance from College:"])

# Normalize the counts to create percentages
ct_pct = ct.apply(lambda r: r/r.sum()*100, axis=1)

# Create a stacked bar chart of the cross-tabulation
ct_pct.plot(kind="bar", stacked=True, figsize=(10, 6))

# Set plot title and axis labels
plt.title("Overall Rating vs Distance from College")
plt.xlabel("Overall Rating")
plt.ylabel("Percentage")

# Show the plot
plt.show()
```

## INFERENCE
### From the stacked bar chart, we can infer that:

* Majority of the hostels with a rating of 5 have a distance from college of less than 500m, and very few have a distance of more than 1km.
* For hostels with a rating of 4 or less, the distribution of distances from college is more spread out.
* Overall, hostels with better ratings tend to be located closer to the college, with a higher proportion having a distance of less than 500m.
